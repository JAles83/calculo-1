# Sección 10: Derivadas de Funciones Especiales y Diferenciación Logarítmica

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import symbols, diff, exp, log, sinh, cosh, tanh, sech, csch, coth, asin, acos, atan, asinh, acosh, atanh, sin, asec, acoth, asech

# Configuración
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

## 1. Derivadas de Funciones Trigonométricas Inversas

**Definiciones clave**:
- $\sin^{-1}x$ o $\arcsin x$: Dominio $[-1, 1]$, Rango $[-\frac{\pi}{2}, \frac{\pi}{2}]$
- $\cos^{-1}x$ o $\arccos x$: Dominio $[-1, 1]$, Rango $[0, \pi]$
- $\tan^{-1}x$ o $\arctan x$: Dominio $\mathbb{R}$, Rango $(-\frac{\pi}{2}, \frac{\pi}{2})$

### Implementación de derivadas

```python
def inverse_trig_derivatives():
    x = symbols('x')
    
    # Ejemplos del PDF
    examples = [
        7*atan(4*x**3),
        2*asec(x**3) - 54*x**7,
        (2*x**6 - x**3*acos(2*x) + 8*sin(3*x**5))
    ]
    
    print("=== Derivadas de Funciones Trigonométricas Inversas ===")
    for i, func in enumerate(examples):
        deriv = diff(func, x)
        print(f"\nEjemplo {i+1}:")
        print(f"Función original: y = {func}")
        print(f"Derivada: y' = {deriv}")
    
    # Gráfico de ejemplo: arctan(x)
    x_vals = np.linspace(-5, 5, 500)
    y = atan(x)
    y_lamb = sp.lambdify(x, y, 'numpy')
    dy_lamb = sp.lambdify(x, diff(y, x), 'numpy')
    
    plt.figure()
    plt.plot(x_vals, y_lamb(x_vals), label='$y = \tan^{-1}x$')
    plt.plot(x_vals, dy_lamb(x_vals), 'r--', label="$y' = \\frac{1}{1+x^2}$")
    plt.title('Función Arco Tangente y su Derivada')
    plt.legend()
    plt.grid(True)
    plt.show()

inverse_trig_derivatives()
```

## 2. Derivadas de Funciones Hiperbólicas

**Definiciones**:
- $\sinh x = \frac{e^x - e^{-x}}{2}$
- $\cosh x = \frac{e^x + e^{-x}}{2}$
- $\tanh x = \frac{\sinh x}{\cosh x}$

### Implementación con visualización

```python
def hyperbolic_derivatives():
    x = symbols('x')
    
    # Ejemplos del PDF
    examples = [
        (9*tanh(x**5)),
        (2*sinh(4*x**6) - 26*x**2),
        (8*x**(-4) + csch(3*x**2) + sinh(x**3)/tanh(2*x**7))
    ]
    
    print("\n=== Derivadas de Funciones Hiperbólicas ===")
    for i, func in enumerate(examples):
        deriv = diff(func, x)
        print(f"\nEjemplo {i+1}:")
        print(f"Función original: y = {func}")
        print(f"Derivada: y' = {deriv}")
    
    # Gráfico comparativo sinh(x) vs cosh(x)
    x_vals = np.linspace(-2, 2, 500)
    sinh_lamb = sp.lambdify(x, sinh(x), 'numpy')
    cosh_lamb = sp.lambdify(x, cosh(x), 'numpy')
    dsinh_lamb = sp.lambdify(x, diff(sinh(x), x), 'numpy')
    dcosh_lamb = sp.lambdify(x, diff(cosh(x), x), 'numpy')
    
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x_vals, sinh_lamb(x_vals), label=r'$\sinh x$')
    plt.plot(x_vals, dsinh_lamb(x_vals), 'r--', label=r"$\frac{d}{dx}\sinh x = \cosh x$")
    plt.title('Derivada del Seno Hiperbólico')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(x_vals, cosh_lamb(x_vals), label=r'$\cosh x$')
    plt.plot(x_vals, dcosh_lamb(x_vals), 'r--', label=r"$\frac{d}{dx}\cosh x = \sinh x$")
    plt.title('Derivada del Coseno Hiperbólico')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

hyperbolic_derivatives()
```

## 3. Derivadas de Funciones Hiperbólicas Inversas

### Implementación con restricciones de dominio

```python
def inverse_hyperbolic_derivatives():
    x = symbols('x')
    
    # Ejemplos del PDF
    examples = [
        (-8*acoth(21*x**3)),
        (6*x**(-4) - acosh(4*x**7)),
        (asech(81*x**4) - 5*x**(-9)*asinh(6*x**7) + 103*x**8)
    ]
    
    print("\n=== Derivadas de Funciones Hiperbólicas Inversas ===")
    for i, func in enumerate(examples):
        deriv = diff(func, x)
        print(f"\nEjemplo {i+1}:")
        print(f"Función original: y = {func}")
        print(f"Derivada: y' = {deriv.simplify()}")
    
    # Gráfico de ejemplo: asinh(x)
    x_vals = np.linspace(-5, 5, 500)
    y = asinh(x)
    y_lamb = sp.lambdify(x, y, 'numpy')
    dy_lamb = sp.lambdify(x, diff(y, x), 'numpy')
    
    plt.figure()
    plt.plot(x_vals, y_lamb(x_vals), label=r'$y = \sinh^{-1}x$')
    plt.plot(x_vals, dy_lamb(x_vals), 'r--', label=r"$y' = \frac{1}{\sqrt{x^2+1}}$")
    plt.title('Función Inversa del Seno Hiperbólico y su Derivada')
    plt.legend()
    plt.grid(True)
    plt.show()

inverse_hyperbolic_derivatives()
```

## 4. Diferenciación Logarítmica

**Técnica útil para**:
- Funciones de la forma $f(x)^{g(x)}$
- Productos/cocientes complejos
- Funciones con variables en exponentes y bases

### Implementación paso a paso

```python
def logarithmic_differentiation():
    x = symbols('x')
    
    # Ejemplo 1 del PDF
    y1 = (log(x)**x)/(2**(3*x+1))
    print("\n=== Diferenciación Logarítmica Ejemplo 1 ===")
    print(f"Función original: y = {y1}")
    
    # Paso 1: Aplicar ln a ambos lados
    ln_y1 = log(y1)
    print(f"\nPaso 1: ln(y) = {ln_y1.simplify()}")
    
    # Paso 2: Diferenciar implícitamente
    deriv_ln = diff(ln_y1, x)
    print(f"\nPaso 2: Derivada implícita: {deriv_ln.simplify()}")
    
    # Paso 3: Resolver para y'
    y1_prime = y1 * deriv_ln
    print(f"\nPaso 3: y' = {y1_prime.simplify()}")
    
    # Ejemplo 2 del PDF (función exponencial anidada)
    y2 = x**(x**(x**4))
    print("\n=== Diferenciación Logarítmica Ejemplo 2 ===")
    print(f"Función original: y = {y2}")
    
    # Aplicamos logaritmo dos veces
    ln_y2 = log(y2)
    ln_ln_y2 = log(ln_y2)
    print(f"\nPaso 1: ln(ln(y)) = {ln_ln_y2.simplify()}")
    
    # Diferenciamos
    deriv_ln_ln = diff(ln_ln_y2, x)
    print(f"\nPaso 2: Derivada implícita: {deriv_ln_ln.simplify()}")
    
    # Resolvemos para y'
    y2_prime = y2 * ln_y2 * deriv_ln_ln
    print(f"\nPaso 3: y' = {y2_prime.simplify()}")

logarithmic_differentiation()
```

## Resumen Teórico Completo

| **Tipo de Función**         | **Derivada Básica**                           | **Con Regla de la Cadena**                     |
|-----------------------------|-----------------------------------------------|-----------------------------------------------|
| **Trig. Inversas**          | $\frac{d}{dx}\sin^{-1}x = \frac{1}{\sqrt{1-x^2}}$ | $\frac{d}{dx}\sin^{-1}u = \frac{u'}{\sqrt{1-u^2}}$ |
|                             | $\frac{d}{dx}\tan^{-1}x = \frac{1}{1+x^2}$       | $\frac{d}{dx}\tan^{-1}u = \frac{u'}{1+u^2}$       |
| **Hiperbólicas**            | $\frac{d}{dx}\sinh x = \cosh x$                  | $\frac{d}{dx}\sinh u = u'\cosh u$                 |
|                             | $\frac{d}{dx}\cosh x = \sinh x$                  | $\frac{d}{dx}\cosh u = u'\sinh u$                 |
| **Hiperb. Inversas**        | $\frac{d}{dx}\sinh^{-1}x = \frac{1}{\sqrt{x^2+1}}$ | $\frac{d}{dx}\sinh^{-1}u = \frac{u'}{\sqrt{u^2+1}}$ |
|                             | $\frac{d}{dx}\tanh^{-1}x = \frac{1}{1-x^2}$       | $\frac{d}{dx}\tanh^{-1}u = \frac{u'}{1-u^2}$       |

**Diferenciación Logarítmica**:
1. Aplicar $\ln$ a ambos lados: $\ln y = \ln f(x)$
2. Diferenciar implícitamente: $\frac{y'}{y} = \frac{d}{dx}[\ln f(x)]$
3. Resolver para $y'$: $y' = y \cdot \frac{d}{dx}[\ln f(x)]$
