# Sección 1:  Definición de Límite

En esta sección exploramos la idea intuitiva del límite de una función, usando ejemplos sencillos para comprender cómo se aproxima el valor de una función cuando la variable independiente se acerca a un punto específico.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
sp.init_printing()
```

## ¿Qué es el límite de una función?

El **límite** de una función es el valor al que la función se aproxima en un valor dado de x, sin importar si la función realmente alcanza ese valor.

Por ejemplo, para la función:

$$
f(x) = x + 1
$$

El límite cuando $x \to 5$ es 6, porque:

$$
\lim_{x \to 5} (x + 1) = 6
$$

```python
# Definimos la función
x = sp.symbols('x')
f = x + 1

# Calculamos el límite cuando x tiende a 5
limite = sp.limit(f, x, 5)

# Mostramos el resultado
print(f"El límite de la función f(x) = {f} cuando x tiende a 5 es: {limite}")
```

### Comprobación gráfica

```python
import numpy as np

# Rango de valores para x alrededor de 5
x_vals = np.linspace(4, 6, 100)
y_vals = x_vals + 1

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='$f(x) = x + 1$')
plt.scatter([5], [6], color='red', zorder=5, label='Punto $(5, 6)$')
plt.axvline(5, color='gray', linestyle='--', alpha=0.5)
plt.axhline(6, color='gray', linestyle='--', alpha=0.5)
plt.title('Visualización del límite de $f(x) = x + 1$ cuando $x \\to 5$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

## Límites laterales

Cuando la función está definida en el punto, el cálculo del límite es tan simple
como introducir el valor de x en la función, es decir:

$$
\lim_{x \to a} f(x) = f(a) \quad \text{si } f(a) \text{ existe}
$$

Sin embargo, cuando la función no esta definida en el punto necesitaremos analizar
los límites laterales.

Un **límite lateral** es el valor al que se aproxima una función cuando la variable independiente se acerca a un punto específico, pero solo desde un lado (izquierda o derecha) de un punto

- Límite por la izquierda:

    Es el valor al que se aproxima $f(x)$ cuando $x$ se acerca a $a$ por valores menores que $a$ (desde la izquierda).
$$ \lim_{x \to a^-} f(x) $$

- Límite por la derecha:
  
    Es el valor al que se aproxima $f(x)$ cuando $x$ se acerca a $a$ por valores mayores que $a$ (desde la derecha).
$$ \lim_{x \to a^+} f(x) $$

```python
# Calculamos los límites laterales
lim_izq = sp.limit(f, x, 5, dir='-')
lim_der = sp.limit(f, x, 5, dir='+')

# Mostramos los límites laterales
print(f"Límite lateral izquierdo cuando x tiende a 5: {lim_izq}")
print(f"Límite lateral derecho cuando x tiende a 5: {lim_der}")
```

## Existencia del límite general

El límite general $\lim_{x \to a} f(x)$ existe si y solo si:

- Existe el límite lateral izquierdo $\lim_{x \to a^-} f(x)$
- Existe el límite lateral derecho $\lim_{x \to a^+} f(x)$
- Ambos límites laterales son iguales

Es decir,
$$
\lim_{x \to a} f(x) = L \iff \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L
$$

```python
# Definimos una función de ejemplo
f = 1 / (x-2)

# Dibujamos la función
x_vals = np.linspace(0, 4, 100)
y_vals = 1 / (x_vals - 2)
plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='$f(x) = \\frac{1}{x - 2}$')
plt.axvline(2, color='red', linestyle='--', label='x = 2')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.ylim(-10, 10)
plt.title('Gráfico de la función $f(x) = \\frac{1}{x - 2}$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.show()
# Calculamos el límite cuando x tiende a 2
limite_2 = sp.limit(f, x, 2)
# Mostramos el resultado del límite
print(f"El límite de la función f(x) = {f} cuando x tiende a 2 es: {limite_2}")
# Calculamos los límites laterales para la función 1/(x-2)
lim_izq_2 = sp.limit(f, x, 2, dir='-')
lim_der_2 = sp.limit(f, x, 2, dir='+')
# Mostramos los límites laterales
print(f"Límite lateral izquierdo cuando x tiende a 2: {lim_izq_2}")
print(f"Límite lateral derecho cuando x tiende a 2: {lim_der_2}")
```

## Definición precisa de límite (definición epsilon-delta)

La definición precisa de límite, conocida como definición epsilon-delta, formaliza el concepto de límite de la siguiente manera:

Sea $f(x)$ una función definida en un intervalo abierto que contiene a $a$, excepto posiblemente en $a$ mismo. Decimos que
$$
\lim_{x \to a} f(x) = L
$$
si, para todo $\varepsilon > 0$, existe $\delta > 0$ tal que, para todo $x$,
$$
0 < |x - a| < \delta \implies |f(x) - L| < \varepsilon
$$

Esto significa que podemos hacer que $f(x)$ esté tan cerca de $L$ como queramos, eligiendo $x$ suficientemente cerca de $a$ (pero distinto de $a$).

**Interpretación:**
- $\varepsilon$ representa una distancia arbitrariamente pequeña respecto al valor límite $L$.
- $\delta$ representa una distancia alrededor de $a$ que garantiza que $f(x)$ esté dentro de la banda de $\varepsilon$ alrededor de $L$.

**Notación:**
$$
\forall \varepsilon > 0,\ \exists\ \delta > 0:\ 0 < |x - a| < \delta \implies |f(x) - L| < \varepsilon
$$

Esta definición es la base rigurosa para el estudio de los límites y el análisis matemático.

```python
# Definimos la función y los parámetros
f = lambda x: 2*x
a = 1
L = 2
epsilon = 0.5

# Calculamos delta para este ejemplo sencillo (f(x) = 2x)
delta = epsilon / 2

# Rango de x
x_vals = np.linspace(0, 2, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label='$f(x) = 2x$')
plt.axvline(a, color='red', linestyle='--', label='$x = a$')
plt.axhline(L, color='green', linestyle='--', label='$L$')

# Banda horizontal epsilon
plt.fill_between(x_vals, L-epsilon, L+epsilon, color='green', alpha=0.2, label='$|f(x)-L|<\\varepsilon$')

# Banda vertical delta
plt.axvline(a-delta, color='blue', linestyle=':', label='$x = a-\\delta$')
plt.axvline(a+delta, color='blue', linestyle=':', label='$x = a+\\delta$')
plt.fill_betweenx(np.linspace(L-epsilon, L+epsilon, 100), a-delta, a+delta, color='blue', alpha=0.1, label='$|x-a|<\\delta$')

plt.scatter([a], [L], color='black', zorder=5)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Visualización de la definición epsilon-delta')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
